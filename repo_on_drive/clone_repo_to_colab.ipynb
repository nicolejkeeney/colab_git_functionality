For cloning a repository from your GitHub. You will need to generate a personal access token from GitHub, and paste it in the variable GIT_TOKEN. See this article for more information: https://towardsdatascience.com/google-drive-google-colab-github-dont-just-read-do-it-5554d5824228

```
MY_GOOGLE_DRIVE_PATH = 'My Drive/github'     # Path to your project on Google Drive

GIT_USERNAME = "nicolejkeeney" 
GIT_REPOSITORY = "colab_git_integration" 
```

```
from google.colab import drive    # Mount Google Drive
ROOT = "/content/drive"           # Default location for the drive
drive.mount(ROOT)                 # Mount the google drive at location specified by ROOT
```

You need to have a personal access key from GitHub in order to clone a repo. My key is stored in a config.py file that is hidden from my repo. You can also just copy paste the token here. Note, I think you only need the token to clone a private repository. The token should match the first part of the url set in the git config file, which is created when you clone the repo. 

```
import sys
sys.path.insert(0, ROOT + '/' + MY_GOOGLE_DRIVE_PATH)

import config
GIT_TOKEN = config.git_token
```

```
# Clone github repository setup
from os.path import join  

PROJECT_PATH = join(ROOT, MY_GOOGLE_DRIVE_PATH)
#print("PROJECT_PATH: ", PROJECT_PATH)   

REPOSITORY_PATH = join(PROJECT_PATH, GIT_REPOSITORY)
#print("REPOSITORY_PATH: ", REPOSITORY_PATH) 

GIT_PATH = "https://" + GIT_TOKEN + "@github.com/" + GIT_USERNAME + "/" + GIT_REPOSITORY + ".git"
#print("GIT_PATH: ", GIT_PATH)
```
